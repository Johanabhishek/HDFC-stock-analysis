```python
df.drop(['Trades', 'Deliverable Volume', '%Deliverble'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

**Feature Engineering**

```python
df['next_close'] = df['Close'].shift(-1)
df.dropna(inplace=True)#removes the last row with nun
```

```python
df['ma_7'] = df['Close'].rolling(window=7).mean()
df['ma_10'] = df['Close'].rolling(window=10).mean()
```

```python
df['daily_return'] = df['Close'].pct_change()
```

```python
df['daily_return']
```

```python
df.drop(['Date', 'Open', 'High', 'Low', 'Volume', 'Symbol','Series'], axis=1, inplace=True)
```

**build and train the model**

```python
from sklearn.model_selection import train_test_split
x=df.drop('next_close', axis=1)
y = df['next_close']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values with the mean
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)
```

```python
imputer = SimpleImputer(strategy='mean')
x_test = imputer.fit_transform(x_test)
```

```python
y_train
```

```python
df.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
y_pred = model.predict(x_test)
print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", mean_squared_error(y_test, y_pred, squared=False))
```

**Data visualization**
